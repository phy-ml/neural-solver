```python
data = pd.read_csv('data.csv')
xcd5 = data.loc[(data['test_fluid']=='XCD5')&(data['annulus']=='annulus_1')]
```

```python
xcd5['shear_pa'] = 13.58
xcd5['k'] = 1.34
```

```python
# convert data into British Units 
# dia (inch) >> feet [add the inner and outer diameters]
# >>>>>> 1 inch = 0.0833333 feet
xcd5['r_out'] = 1.38 * 0.0833333
xcd5['r_in'] = 0.38 * 0.0833333
#xcd5
```

```python
# flowrate(gpm) >> m3/s
# >>>>> 1GPM = 0.00222801 ft3/s
xcd5['flow_rate_gpm'] = xcd5['flow_rate_gpm']*0.00222801
#xcd5
```

```python
# diff press (inchH20) >> KPa/meter
# convert inch h20 into PSI ( 1 inchH20 = 0.0360912 PSI )
xcd5['diff_p'] = (5.20/18)*xcd5['diff_p'] # for Lbf/ft2/ft
xcd5
```

```python
plt.scatter(xcd5['flow_rate_gpm'], xcd5['diff_p'], color='black', label='Exp Data')
plt.grid('both')
plt.xlabel('f3/s')
plt.ylabel('Psi/ft')
plt.legend();
```
